```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('dataset1.csv')
df2 = pd.read_csv('dataset2.csv')
df3 = pd.read_csv('dataset3.csv')
```

```python
df1.head()
```

```python
import plotly.express as px
fig = px.scatter_3d(df1, x='F0', y='F1', z='F2',
              color='F3', symbol='LABEL')
fig.show()
```

```python
fig = px.scatter_3d(df2, x='F0', y='F1', z='F2',
              color='F3', symbol='LABEL')
fig.show()
```

```python
fig = px.scatter_3d(df3, x='F0', y='F1', z='F2',
              color='F3', symbol='LABEL')
fig.show()
```

```python
from sklearn.cluster import KMeans
```

```python
clst = KMeans(n_clusters=3, random_state=42).fit(df1.drop('LABEL', axis=1))
```

```python
df1['clst'] = clst.labels_
```

```python
fig = px.scatter_3d(df1, x='F0', y='F1', z='F2',
              color='F3', symbol='clst')
fig.show()
```

```python
clst.cluster_centers_
```

```python
kmeans = KMeans(n_clusters=2, random_state=42).fit(df2.drop('LABEL', axis=1))
df2['clst'] = kmeans.labels_
px.scatter_3d(df2, x='F0', y='F1', z='F2', color='F3', symbol='clst')
```

```python
from sklearn.cluster import AgglomerativeClustering

AgglClust = AgglomerativeClustering(linkage='complete').fit(df2.drop('LABEL', axis=1))
df2['agglclst'] = AgglClust.labels_
px.scatter_3d(df2, x='F0', y='F1', z='F2', color='F3', symbol='agglclst')
```

```python
df2['agglclst'].value_counts()
```

# **DBSCAN**

```python
from sklearn.cluster import DBSCAN
```

```python
dbscn = DBSCAN(eps=1.5, min_samples=8).fit(df3.drop('LABEL', axis=1))
df3['clst'] = dbscn.labels_
px.scatter_3d(df3, x='F0', y='F1', z='F2', color='F3', symbol='clst')
```

```python
df3['clst'].value_counts()
```

```python
dbscn = DBSCAN(eps=0.3, min_samples=10).fit(df3.drop('LABEL', axis=1))
df3['clst_0'] = dbscn.labels_
px.scatter_3d(df3, x='F0', y='F1', z='F2', color='F3', symbol='clst_0')
```
