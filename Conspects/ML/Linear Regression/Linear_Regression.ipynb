```python
df = pd.read_csv('poly.csv')
df.head()
```

```python
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures


X_train, X_test, y_train, y_test = train_test_split(df['X'],
                                                    df['Y'],
                                                    test_size=0.2,
                                                    random_state=10)
X_train = np.array(X_train).reshape(-1, 1)
X_test = np.array(X_test).reshape(-1, 1)


scores = []
for alph in range(1, 11):
    for i in range(1, 8):
        linreg = make_pipeline(
            PolynomialFeatures(degree=i),
            Ridge(alpha = alph)
        )
        linreg.fit(X_train, y_train)
        y_pred = linreg.predict(X_test)
        scores.append((alph, i, r2_score(y_pred, y_test)))
```

```python
df = pd.DataFrame(scores, columns=['alpha', 'degree', 'r2_score'])
df.head()
```
